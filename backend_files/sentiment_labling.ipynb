# Sentiment Labling to the reviews

```python
from textblob import TextBlob
import pandas as pd
import streamlit as st
import cleantext

def sentiment_analyze(x):
    
    blob1 = TextBlob(str(x))
    score = blob1.sentiment.polarity
    
    if score >= 0.5:
        return 'Positive'
    elif score <= -0.5:
        return 'Negative'
    else:
        return 'Neutral'


file = r'C:\Users\itw\Downloads\all csv\merged_reviews.csv'
df = pd.read_csv(file)
df['sentiment'] = df['review'].apply(sentiment_analyze)
```

```python
df
labled_file = r'C:\Users\itw\Downloads\all csv\labled_data.csv'
df.to_csv(labled_file, index=False)
```
