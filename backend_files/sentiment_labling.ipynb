# Sentiment Labling to the reviews

```python
from textblob import TextBlob
import pandas as pd
import streamlit as st
import cleantext

def sentiment_analyze(x):
    
    blob1 = TextBlob(str(x))
    score = blob1.sentiment.polarity
    
    if score >= 0.5:
        return 'Positive'
    elif score <= -0.5:
        return 'Negative'
    else:
        return 'Neutral'


file = r'C:\Users\itw\Downloads\all csv\merged_reviews.csv'
df = pd.read_csv(file)
df['sentiment'] = df['review'].apply(sentiment_analyze)
```

# Scoring the reviews


```python
from textblob import TextBlob
import pandas as pd

def score_analyze(x):
    
    blob1 = TextBlob(str(x))
    score = blob1.sentiment.polarity
    return 3.0 + (score * 3.0)


file = r'C:\Users\itw\Downloads\all csv\labled_data.csv'
df = pd.read_csv(file)
df['score'] = df['review'].apply(score_analyze).round(1)
```

# Finding Average Review Score of all hotels

```python
hotel_scores = df.groupby('hotel_name')['score'].mean().reset_index().round(1)
hotel_scores.rename(columns={'score': 'review_score'}, inplace=True)
hotel_scores
```

# Merging it into the hotel_info File

```python
file_hotel_info = r'C:\Users\itw\Downloads\all csv\merged_hotel_info.csv'
df_hotel_info = pd.read_csv(file_hotel_info)

# Ensure hotel_name columns are stripped of whitespace and converted to lowercase
df_hotel_info['hotel_name'] = df_hotel_info['hotel_name'].str.strip().str.lower()
hotel_scores['hotel_name'] = hotel_scores['hotel_name'].str.strip().str.lower()

# Merge the hotel_info DataFrame with the hotel_scores DataFrame
df_hotel_info = df_hotel_info.merge(hotel_scores, on='hotel_name', how='left')
```
