# Below is the cleaning of islamabad hotels file

```python
import pandas as pd

# Read the CSV file
file_path = r'C:\Users\itw\Downloads\islamabad csv\islamabad_hotel_names.csv'
file_path1 = r'C:\Users\itw\Downloads\islamabad csv\islamabad_hotel_names_corrected.csv'

df = pd.read_csv(file_path)

# Rename columns as per the requirement
df.columns = ['hotel_link', 'hotel_name', 'review_score', 'total_reviews', 'hotel_type', 'hotel_price']

# Keep only specific rows (1, 7, 5, 8, 10)
rows_to_keep = [1, 7, 5, 8, 10]
df = df.iloc[rows_to_keep]

# Remove '-' from values in the 'total_reviews' column
df['total_reviews'] = df['total_reviews'].str.replace('-', '')

# Remove 'PKRÂ ' from values in the 'hotel_price' column
df['hotel_price'] = df['hotel_price'].str.replace('PKRÂ ', '')

# Write the modified DataFrame back to the same CSV file
df.to_csv(file_path1, index=False)
```

# Cleaning of all review files of lahore

```python
files = ['\Faletti hotel lahore reviews.csv', '\lahore avari hotels reviews.csv', '\luxus grand hotel lahore reviews.csv', '\Pearl Continental lahore Reviews.csv', '\The Nishat Hotel reviews lahore.csv']
path = r'C:\Users\itw\Downloads\lahore csv'
corrected_files = ['\Faletti hotel lahore reviews corrected.csv', '\lahore avari hotels reviews corrected.csv', '\luxus grand hotel lahore reviews corrected.csv', '\Pearl Continental lahore Reviews corrected.csv', '\The Nishat Hotel reviews lahore corrected.csv']

i=0
for file in files:
    file_path = f'{path}\\{file}'
    df = pd.read_csv(file_path)

# Remove first and third columns
    df.drop(df.columns[[0, 2]], axis=1, inplace=True)

    # Rename remaining columns as per the requirement
    df.columns = ['reviewer_name', 'review_time', 'review']
    
    file_path1 = f'{path}\\{corrected_files[i]}'
    # Write the modified DataFrame back to the same CSV file
    i = i + 1
    df.to_csv(file_path1, index=False)
```
